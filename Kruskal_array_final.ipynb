```python
def kruskal_array(G): 
    print('FromNode, ToNode, Distance')
    minimum = 0 # Cost of min MST 
  
    # Initialize sets of disjoint sets 
    for i in range(V): 
        parent[i] = i 
  

    edges = 0
    #initiate edge count at 0
    #goes through each edge one by one
    while edges < (V - 2): 
        min = INF 
        node1 = -1
        node2 = -1
        for i in range(V): 
            for j in range(V): 
                if find(i) != find(j) and G[i][j] < min: 
                    min = G[i][j] 
                    node1 = i 
                    node2 = j 
        
        union(node1, node2) 
        print( node1,',',     node2,',',      min) 
        edges = edges+ 1
        #add one to edge count
        minimum += min
  
    print('Total distance') 
    print(minimum)

def union(i, j): 
    k = find(i) 
    l = find(j) 
    parent[k] = l 

def find(i): 
    while parent[i] != i: 
        i = parent[i] 
    return i 
  

  
```
